```python
%matplotlib inline

from pymote3.algorithms.broadcast import Flood
from pymote3 import NetworkGenerator
from pymote3 import write_pickle

net_gen = NetworkGenerator(100)
net = net_gen.generate_random_network()
net.show()

net.algorithms = ((Flood, {'informationKey': 'I'}), )

ini_node = net.nodes_list()[0]
ini_node.memory['I'] = 'Hello World'

write_pickle(net, "Flood.npc.gz")
```

```python
import numpy as np

net_gen = NetworkGenerator(100)
net = net_gen.generate_random_network()
net.show()

net.algorithms = ((Flood, {'informationKey': 'I'}), )

for n in net.nodes():
    if np.random.random() < 0.25:
        n.memory['I'] = 'MultiFlood'
    
write_pickle(net, "MultiFlood.npc.gz")
```

```python
from pymote3 import NetworkGenerator
from pymote3 import write_pickle
from pymote3.algorithms.dft import DFT

net_gen = NetworkGenerator(10)
net = net_gen.generate_random_network()
net.show()

net.algorithms = (DFT, )

write_pickle(net, "DFT.npc.gz")
```
