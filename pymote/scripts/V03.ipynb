Every algorithm instance has a set of required and default params:
* Required params must be given to algorithm initializer as a keyword arguments.
* Default params can be given to algorithm initializer as a keyword arguments, if not their class defines default value.

```python
%matplotlib inline

from pymote.algorithms.broadcast import Flood
from pymote import NetworkGenerator
from pymote import write_pickle

net_gen = NetworkGenerator(100)
net = net_gen.generate_random_network()
net.show()

net.algorithms = ((Flood, {'informationKey': 'I'}), )

ini_node = net.nodes()[0]
ini_node.memory['I'] = 'Hello World'

write_pickle(net, "Flood.npc.gz")
```

```python
import numpy as np

net_gen = NetworkGenerator(100)
net = net_gen.generate_random_network()
net.show()

net.algorithms = ((Flood, {'informationKey': 'I'}), )

for n in net.nodes():
    if np.random.random() < 0.25:
        n.memory['I'] = 'MultiFlood'
    
write_pickle(net, "MultiFlood.npc.gz")
```

```python
from pymote import NetworkGenerator
from pymote import write_pickle
from pymote.algorithms.dft import DFT

net_gen = NetworkGenerator(10)
net = net_gen.generate_random_network()
net.show()

net.algorithms = (DFT, )

write_pickle(net, "DFT.npc.gz")
```

```python
from pymote import read_pickle

net_file = read_pickle('DFT.npc.gz') 
net_file.show()
```

```python
from pymote import Simulation

sim = Simulation(net_file)
```

```python
sim.reset()
sim.run()
```
